```python
from pyclausie import ClausIE
from tqdm import tqdm
from nltk.tokenize import sent_tokenize
import os
import re
import nltk
nltk.download('punkt')
from os import walk
from tqdm.notebook import tqdm
```

```python

def all_clausIE():
    # Get all files:
    fns = next(walk('/content/drive/MyDrive/ColabNotebooks/MPR1/Files/Files_txt/'), (None, None, []))[2]  # [] if no file

    # Create dictionary for counter
    pred_counter = {}

    for filename in fns:
      pred_counter = indv_clausIE(filename, pred_counter)

      save_dict(pred_counter)
      #break

    return pred_counter

def save_dict(pred_counter):
  # Save everything
  dict_pd_words = []
  for k, v in sorted(pred_counter.items(),key = lambda item: -item[1]):
      dict_pd_words.append([k,v])

  import pandas as pd
  dict_pd = pd.DataFrame(dict_pd_words, columns=[['keys', 'count']])
  print(dict_pd.head())

  dict_pd.to_csv("/content/drive/MyDrive/ColabNotebooks/MPR1/Files/dictionary_counter.csv")


def indv_clausIE(filename, pred_counter):

    # Create clausIE instance
    cl= ClausIE.get_instance()

    # Read document
    with open('/content/drive/MyDrive/ColabNotebooks/MPR1/Files/Files_txt/' + filename, encoding="utf-8") as f:
      lines=f.readlines()

    # Get all paragraphs and remove enters
    list = [x.rstrip() for x in lines]

    counter = 0

    # For each paragraph
    for par in tqdm(list):
      # For each sentence in paragraph
      for s in sent_tokenize(par):
          # Get triples using clausie
          # TODO - multi-thread? if-statement?

          try:

            triples = cl.extract_triples([s])

            # For each triple
            for triple in triples:
              # For each predicate loop over all word
              for pred in triple.predicate.split(" "):
                  # if words exist add 1, else create new
                  if pred in pred_counter.keys():
                      pred_counter[pred] += 1
                  else:
                      pred_counter[pred] = 1

          except Exception:
            print('error jump')
            continue
      counter += 1

      #if counter > 100:
      #  break

    # Print sorted results
    #for k, v in sorted(pred_counter.items(),key = lambda item: -item[1]):
    #  print(k, v)

    return pred_counter
```
